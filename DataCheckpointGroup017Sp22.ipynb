```python
aqi.head()
aqi_ca = aqi[aqi['state'] == 'California']
aqi_ca = aqi_ca.reset_index(drop=True)
aqi_ca.drop(['good days', 'moderate days', 'very unhealthy days', 'unhealthy days',
                'hazardous days', 'unhealthy for sensitive groups days'],
               inplace = True, axis=1)
#aqi_ca['county'] = aqi_ca['county'].apply(standardize_region)

#aqi_ca.shape
county_cancer_2010 = county_cancer[county_cancer['Year'] == 2010]
county_cancer_2010 = county_cancer_2010[county_cancer_2010['Age'] == 'All Ages']
county_cancer_2010['California Region'] = county_cancer_2010['California Region'].apply(standardize_region)
county_cancer_2010
```

```python
# remove " county"
def standardize_region(str_in):
    
    try:
        str_in = str_in.lower()
        str_in = str_in.strip()

        if 'county' in str_in:
            str_in = str_in.replace('county', '')
            output = str_in
    except:
        output = np.nan
        
    return output
```
